```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/HR-Employee-Attrition.csv')

# Data Cleaning

# Deleting redundant columns
redundant_columns = ['EmployeeCount', 'EmployeeNumber', 'StandardHours', 'Over18']
data = data.drop(columns=redundant_columns)

# Dropping duplicates
data = data.drop_duplicates()

# Cleaning individual columns (if needed)
# Example: data['column_name'] = data['column_name'].apply(clean_function)

# Remove the NaN values from the dataset
data = data.dropna()

# Check for some more Transformations
# Example: data['column_name'] = data['column_name'].apply(transformation_function)

# Data Visualization

# Plot a correlation map for all numeric variables
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Map")
plt.show()

# Overtime
sns.countplot(x='OverTime', data=data)
plt.title("Overtime")
plt.show()

# Marital Status
sns.countplot(x='MaritalStatus', data=data)
plt.title("Marital Status")
plt.show()

# Job Role
plt.figure(figsize=(10, 6))
sns.countplot(x='JobRole', data=data, hue='Attrition')
plt.xticks(rotation=90)
plt.title("Job Role vs. Attrition")
plt.show()

# Gender
sns.countplot(x='Gender', data=data, hue='Attrition')
plt.title("Gender vs. Attrition")
plt.show()

# Education Field
plt.figure(figsize=(10, 6))
sns.countplot(x='EducationField', data=data, hue='Attrition')
plt.xticks(rotation=90)
plt.title("Education Field vs. Attrition")
plt.show()

# Department
sns.countplot(x='Department', data=data, hue='Attrition')
plt.title("Department vs. Attrition")
plt.show()

# Business Travel
sns.countplot(x='BusinessTravel', data=data, hue='Attrition')
plt.title("Business Travel vs. Attrition")
plt.show()

# Relation between Overtime and Age
sns.boxplot(x='OverTime', y='Age', data=data)
plt.title("Relation between Overtime and Age")
plt.show()

# Total Working Years
sns.histplot(data=data, x='TotalWorkingYears', hue='Attrition', bins=20, kde=True)
plt.title("Total Working Years vs. Attrition")
plt.show()

# Education Level
sns.countplot(x='Education', data=data, hue='Attrition')
plt.title("Education Level vs. Attrition")
plt.show()

# Number of Companies Worked
sns.histplot(data=data, x='NumCompaniesWorked', hue='Attrition', bins=10, kde=True)
plt.title("Number of Companies Worked vs. Attrition")
plt.show()

# Distance from Home
sns.histplot(data=data, x='DistanceFromHome', hue='Attrition', bins=20, kde=True)
plt.title("Distance from Home vs. Attrition")
plt.show()
```
